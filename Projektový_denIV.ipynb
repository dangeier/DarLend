# Více vstupů

vice_neuronu.png

```
inputs = [1, 2, 3, 2.5]
weights = [[0.2, 0.8, -0.5, 1], [0.5, -0.91, 0.26, -0.5], [-0.26, -0.27,0.17, 0.87]]
bias = [2, 3, 0.5]

list_of_outputs = []
for i in range(len(weights)):
  list_of_outputs.append(sum([ inp * weight for inp, weight in zip(inputs, weights[i])]) + bias[i])
print(list_of_outputs)
```

# Vytvoreni site


image.pnghttps://missinglink.ai/wp-content/uploads/2018/11/activationfunction-1.png

```
network = models.Sequential()
network.add(layers.Dense(512, activation='relu', input_shape=(28 * 28,))) #layer=vrstva 
#datovy fitr... data vstoupi a vystoupi v uyitecne forme.
#vrstvy extrahuji reprezentace
network.add(layers.Dense(10, activation='softmax')) #posledni vrstve vraci pole 10 prvku
#jedna se o klasifikaci 10 trid!
#každá třída bude zastoupena prvděpodobností výskytu danného čísla na obrázku. soucet = 1
```

# Relu

image.png

```
list_to_next_layer = []
args = [-5,6,-1,0,2,3,5] 
for x in args:
  if x <= 0:
    x = 0
    list_to_next_layer.append(x) 
  else:
    x = x
    list_to_next_layer.append(x)  
print(list_to_next_layer)
```

# Kompilační krok


```
network.compile(optimizer='rmsprop', #timto mechanismem se sit aktualizuje
                loss='categorical_crossentropy', #Jak bude sit schopna merit vykon na trenovacich datech
                # a tim i na testovacich
                metrics=['accuracy'])#podil obrazku, ktere byli pri trenovani spravne urcene a ktere ne!
```

#Příprava obrazových dat
- upravime data do intervalu 0,1
- předtím jsme měli: (60000,28,28) typu uint8 v intervalu [0-255] 
- prevedeme do float32

```
train_images = train_images.reshape((60000, 28 * 28))
train_images = train_images.astype('float32') / 255

test_images = test_images.reshape((10000, 28 * 28))
test_images = test_images.astype('float32') / 255
```

```
#Priprava oznaceni trid.
from keras.utils import to_categorical
n_classes = 10
train_labels = to_categorical(train_labels,n_classes)
test_labels = to_categorical(test_labels,n_classes)
```

```
train_labels[0]
```

```
train_labels[5]
```

#Trenovani

```
network.fit(train_images, train_labels, epochs=5, batch_size=128)#5 epoch po davkach 128
```

# loss_funcion
Funkce ztráty je jednou z důležitých součástí neuronových sítí. Ztráta není nic jiného než chyba předpovědi Neural Net. A metoda výpočtu ztráty se nazývá funkce ztráty.
Jednoduše řečeno, ztráta se používá k výpočtu gradientů. A přechody se používají k aktualizaci vah neurální sítě. Takto se trénuje neurální síť.

#Accurncy

```
#Ztrata na trenovacich datech  a správnost sítě na trénovacích datech
test_loss, test_acc = network.evaluate(test_images, test_labels)  #zjisti stratu a presnost
#Pri kazde iteraci sit vypocte gradienty vah s ohledem na davku a podle toho se vahy aklualizuji.
```
